```python
def get_layer(gt, label_list):
    result = []
    for i in label_list:
        if i in gt:
            result.append(np.argwhere(gt == i)[0][0])
    if len(result) == 0:
        result.append(-1)
    return result
```

```python
def create_layered_labels(row):
    img_name = row[0]
    labelname = row[1]
    labels = []
    labelname_list = labelname.split()
    
    layer_one = get_layer(layer_one_gt, labelname_list)
    layer_two = get_layer(layer_two_gt, labelname_list)
    layer_three = get_layer(layer_three_gt, labelname_list)
    layer_four = get_layer(layer_four_gt, labelname_list)
    
    record = [img_name, layer_one, layer_two, layer_three, layer_four]
    
    return record
```
